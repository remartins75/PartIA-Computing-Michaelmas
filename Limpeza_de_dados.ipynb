```python
#####Passo 1 = importar base de dados para limpeza##########

#importar biblioteca que permite a importacao de tabela
import pandas as pd
import re

#busca do arquivo para tratativa
tabela = pd.read_excel("/content/drive/MyDrive/Classroom/Projeto_limpeza de base/DRE 2019.xlsx")
#display(tabela)
# Exibe as primeiras linhas da coluna 'nome da coluna'

# Verificar as colunas para verificar e garantir os nomes da primeira linha (nome das colunas) estão corretas:
#print(tabela.columns)

#######Passo 2 = visualizar e tratar a base de dados e efetuar tratativas##################
"""
.fillna("") - utilizado para substituir qualquer informacao (null | NaN) em 'branco';
.astype(str) - deixar todos os textos garantidos como string
str.upper() - em letra maiúscula
str.replace(" ", "", regex=False) sendo utilizado o regex=false para garantir a substituicao completa do replace
para uma informacao regular sem qualquer problema. É uma opcao mais avancada do replace.
display() - utilizar o display() em vez do print() permite trazer as informacoes da tabela em um formato mais visual e correto no python
print(tabela.colums) - verificar se os nomes das primeras linhas estavam corretas para transcrever no código
tabela.to_excel("caminho_alteracao do nome final)"- exportar o arquivo tratado
"""

# Limpezas e Garantias

#limpeza de dados [Nome]
tabela['Nome'] = tabela['Nome'].fillna("").astype(str).str.upper().str.replace(" ", "", regex=False)

# Garantir que 'Endereco' é tratado como string e depois aplicar as transformações | transformando a coluna em string
tabela['Endereco'] = tabela['Endereco'].fillna("").astype(str).str.upper().str.replace(" ", "", regex=False)

# Garantir que 'Descrição da Conta' é tratado como string, substituindo NaN por string vazia e depois aplicar as conversoes
tabela['Descrição da Conta'] = tabela['Descrição da Conta'].fillna('').astype(str).str.upper().str.replace(" ", "", regex=False)

# Garantir que 'Nível 2' é tratado como string, substituindo NaN por string vazia e depois aplicar as transformações
tabela['Nível 2'] = tabela['Nível 2'].fillna("").astype(str).str.upper().str.replace(" ", "", regex=False)

# Garantir que 'Nível 1' é tratado como string, substituindo NaN por string vazia e depois aplicar as transformações
tabela['Nível 1'] = tabela['Nível 1'].fillna('').astype(str).str.upper().str.replace(" ", "", regex=False)

display(tabela)

## CONFERÊNCIA: Exibir as primeiras linhas da tabela limpa | entre parenteses ajustar a quantidade de linhas se necessário
#display(tabela.head(30))

##########Passo 3 = exportar a base de dados tratada com os campos necessarios##################

tabela.to_excel("/content/drive/MyDrive/Classroom/Projeto_limpeza de base/DRE 2019_TRATADO.xlsx")






```
